# Activation Functions

## Relu function

The Relu activation functions is the most used one, $\text{ReLU}(x) = \max(0, x)$.

The intuition behind the Relu activation function is that if the input x is negative the output will be 0 and if the input is positive just take it as output.

The reason why it's commonly used is its derivative which behave well on optimization $\text{ReLU}'(x) = 0$ if x < 0 and $\text{ReLU}'(x) = 1$ if x > 0, note that Relu is not differentiable for x = 0 but usually we take $\text{ReLU}'(x) = 0$.

There exist other variations of the Relu activation function like $\text{PReLU}(x) = \max(0, x) + \alpha \min(0, x)$.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return max(0,x)

def relu_derivative(x):
    return np.where(x <= 0, 0, 1)

def plot_relu():
    x = np.linspace(-10, 10, 100)
    
    relu = np.maximum(0, x)
    relu_deriv = relu_derivative(x)

    plt.figure(figsize=(10, 4))

    # ReLU function
    plt.subplot(1, 2, 1)
    plt.plot(x, relu, label="ReLU")
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.title('ReLU Activation Function')
    plt.grid(True)
    plt.legend()

    # ReLU derivative
    plt.subplot(1, 2, 2)
    plt.plot(x, relu_deriv, label="ReLU Derivative", color='r')
    plt.xlabel('Input')
    plt.ylabel('Derivative')
    plt.title('ReLU Derivative')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_relu()
```

## Sigmoid function

$\text{sigmoid}(x): \mathbb{R} \to [0,1]$, it takes an input from $\mathbb{R}$ and as output we get y in the range $[0,1]$, this is why often it's called a *squashing function*.

$$ 
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Generally used as activation function of the output layer of an artificial neural network for binary classification (we interpret the result as as probability for example if sigmoid(x) = 0.7 then x has 70% chances being in class 1). We can consider the sigmoid function as a special case of the softmax activation function.

Sigmoid is diffirentiable in $\mathbb{R}$ and its derivative:
$$
\sigma'(x) = \frac{\exp(-x)}{(1 + \exp(-x))^2} = \sigma(x) (1 - \sigma(x))
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def plot_sigmoid_and_derivative():
    x = np.linspace(-10, 10, 100)
    y = sigmoid(x)
    y_derivative = sigmoid_derivative(x)

    fig, axs = plt.subplots(1, 2, figsize=(12, 4))

    axs[0].plot(x, y, label="Sigmoid", color="blue")
    axs[0].set_xlabel("x")
    axs[0].set_ylabel("y")
    axs[0].set_title("Sigmoid Function")
    axs[0].legend()
    axs[0].grid()

    axs[1].plot(x, y_derivative, label="Sigmoid Derivative", color="red")
    axs[1].set_xlabel("x")
    axs[1].set_ylabel("y")
    axs[1].set_title("Sigmoid Derivative")
    axs[1].legend()
    axs[1].grid()

    plt.show()

plot_sigmoid_and_derivative()
```

Note than when x approaches 0, the sigmoid function approaches a linear transformation
